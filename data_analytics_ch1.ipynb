# Nock4 
複数データをUnion してみる

```python
order_all = pd.DataFrame()
file = tbl_order_files[0]
order_data = pd.read_csv(file)

print(f'{file}:{len(order_data)}')
order_all = pd.concat([order_all, order_data], ignore_index=True)
order_all
```

```python
order_all = pd.DataFrame()
for file in tbl_order_files: 
    order_data = pd.read_csv(file)
    print(f'{file}:{len(order_data)}')
    order_all = pd.concat([order_all, order_data],ignore_index=True)
```

```python
order_all
```

# Nock 6: 不要なデータを除外する

```python
order_data = order_all.loc[order_all['store_id']!= 999]
order_data
```

# Nock 7: Join with Master Data

```python
#print(m_store)
order_data = pd.merge(order_data, m_store, on='store_id', how='left')
order_data
```

```python
order_data = pd.merge(order_data, m_area, on='area_cd', how='left')
order_data
```

# Nock8: Name codes that are not in master data

```python
# in this case, we will name for "takeout_flag" and "status"
# Case: takeout_flag == 0
order_data.loc[order_data['takeout_flag'] == 0, 'takeout_name'] = 'デリバリー'
# Case: takeout_flag == 1 
order_data.loc[order_data['takeout_flag'] == 1, 'takeout_name'] = 'お持ち帰り'
order_data
```

```python
# Case: status == 0
order_data.loc[order_data['status'] == 0, 'status'] = '受付'
# Case: status == 1 
order_data.loc[order_data['status'] == 1, 'status'] = 'お支払い済み'
# Case: status == 2
order_data.loc[order_data['status'] == 2, 'status'] = 'お渡し済み'
# Case: status == 9
order_data.loc[order_data['status'] == 9, 'status'] = 'キャンセル'
order_data
```

# Nock 9: 分析基礎テーブルの出力

```python
output_file = os.path.join(output_dir, 'order_data.csv')
order_data.to_csv(output_file, index=False)
```
